```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv')

plt.figure(figsize = (16, 16), constrained_layout=True) # figure size
# plt.tight_layout()

plt.subplot(221)
x = df["Township"]
male = df["Male"]
female = df["Female"]
total =  df["Total(Male+Female)"]
plt.plot(x, male, label="Male")
plt.plot(x, female, label="Female", linestyle="dashed")
plt.plot(x, total, label="Total(Male+Female)", linestyle="dashed")
index = np.arange(len(df.Township))
plt.xticks(index, df.Township,  rotation = 60)
# Label the axis
plt.title("Line graph showing population")
plt.xlabel("Township")
plt.ylabel("Number of people")
# Add legend
plt.legend()


plt.subplot(222)
x = df["Township"]
immigration = df["Immigration"]
emigration = df["Emigration"]
plt.plot(x, immigration, label="Immigration")
plt.plot(x, emigration, label="Emigration", linestyle="dashed")
index = np.arange(len(df.Township))
plt.xticks(index, df.Township,  rotation = 60)
# Label the axis
plt.title("Line graph showing immigration and emigration")
plt.xlabel("Township")
plt.ylabel("Number of people")
# Add legend
plt.legend()


plt.subplot(223)
x = df["Township"]
y = df["Number_of_Households"]
plt.bar(x, y)
index = np.arange(len(df.Township))
plt.xticks(index, df.Township,  rotation = 60)
#Label the axis
plt.title("Bar graph showing number of households")
plt.xlabel("Township")
plt.ylabel("Number of people")


plt.subplot(224)
x = df["Township"]
birth = df["Birth"]
death = df["Death"]
plt.plot(x, birth, label = "Birth")
plt.plot(x, death, linestyle='dashed', label = "Death")
index = np.arange(len(df.Township))
plt.xticks(index, df.Township,  rotation = 60)
# Label the axis
plt.title("Bar graph showing births and deaths")
plt.xlabel("Township")
plt.ylabel("Number of people")
# Add legend
plt.legend()  

plt.suptitle("Miaoli County population data", fontsize=18, fontweight='bold')
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("data.csv")
df[::5]
```

```python
output_df = df.groupby(by=['High school system', 'Gender'], as_index=False).sum()
output_df = output_df[["High school system", "Gender", "Num of graduates", "Num of enroll in police university", "Num of enroll in police college"]]
output_df["Military schoool enrollment average"] = df.groupby(by=['High school system', 'Gender'], as_index=False).mean()['Num of enroll in military school']
output_df["Military schoool enrollment average"]  = output_df["Military schoool enrollment average"].map('{:,.4f}'.format)

output_df["Police university enrollment rate"] = (output_df["Num of enroll in police university"] / output_df["Num of graduates"]).round(4)
output_df["Police college enrollment rate"] = (output_df["Num of enroll in police college"] / output_df["Num of graduates"]).round(4)
output_df["Millitary school enrollment rate"] = ( df.groupby(by=['High school system', 'Gender'], as_index=False).sum()["Num of enroll in military school"] / output_df["Num of graduates"]).round(4)
```
